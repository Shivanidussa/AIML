```python
# uppercase,lowercase,special characters,numbers

#uppercase A - Z
uppercase = random.sample('ABCDEFGHIJKLMNOPQRSTUVWXYZ',random.randint(1,3))
print(uppercase)

# lowercase a - z
lowercase = random.sample('abcdefghijklmnopqrstuvwxyz',random.randint(1,4))
print(lowercase)

#numbers 0 - 10
numbers = random.sample('0123456789',random.randint(0,9))
print(numbers)

#special Charcaters !@#$%^&*/
spcl = random.sample('!@#$%^&*/',random.randint(2,5))
print(spcl)
```

```python
def createPswrd():
    uppercase = random.sample('ABCDEFGHIJKLMNOPQRSTUVWXYZ',random.randint(1,3))
    lowercase = random.sample('abcdefghijklmnopqrstuvwxyz',random.randint(1,4))
    numbers = random.sample('0123456789',random.randint(0,9))
    spcl = random.sample('!@#$%^&*/',random.randint(2,5))
    password = uppercase + lowercase + numbers + spcl 
    np.random.shuffle(password)
    pwd = ''.join(password)
    return pwd
```

```python
pwrd_created = createPswrd()
print(pwrd_created)
```
