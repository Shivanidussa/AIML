```python
def rectifyproductType(pd):
    if pd == 'coates':
        return 'Coats/Jackets'
    elif pd == 'dresses':
        return 'Dresses'
    elif pd == 'dresses':
        return 'Dresses'
    else:
        return pd
    
def rectifyRating(rt):
        if rt == '1 start':
            return '1 star'
        elif rt == '2star':
            return '2 star'
        else:
            return rt
        
        
def rectifypaymentmode(pyt):
        if pyt == 'Creditcard':
            return 'Credit Card'
        elif pyt == 'Directdeposit':
            return 'Direct Deposit'
        else:
            return pyt    
```

```python
bellofshn.ProductType = bellofshn.ProductType.apply(rectifyproductType)
bellofshn.Rating = bellofshn.Rating.apply(rectifyRating)
bellofshn.PaymentMode = bellofshn.PaymentMode.apply(rectifypaymentmode)
```

```python
def getcurrency(price):
    if pd.isnull(price):
        return ''
    elif '$' in price:
        return 'US dollar'
    elif '£' in price:
        return 'Pounds'
    elif in price:
        return 'Euros'
    elif '₹' in price:
        return  '€''Indian Rupees'
    else:
        return ''
```

```python
bellofshn['PaidCurrency'] = bellofshn.Price.apply(getcurrency)
bellofshn.head()
```

```python
bellofshn['Price_USD'] = bellofshn.Price
bellofshn[['Date','Price','PaidCurrency','Price_USD']]
```

```python
bellofshn.Price_USD = bellofshn.Price_USD.str.replace('$','')
bellofshn.Price_USD = bellofshn.Price_USD.str.replace('£','')
bellofshn.Price_USD = bellofshn.Price_USD.str.replace('€','')
bellofshn.Price_USD = bellofshn.Price_USD.str.replace('₹','')
bellofshn.Price_USD = bellofshn.Price_USD.str.replace(',','')
bellofshn[['Date','Price','PaidCurrency','Price_USD']]
```

```python
bellofshn[bellofshn.Price_USD.isnull()]
```

```python
erate =[1.171816, 1.367325, 0.013544]
currency = ['€', '£' , '₹']
```

```python
print(erate)
print(bellofshn.PaidCurrency.unique())
```

```python
bellofshn.reset_index(inplace = True, drop = True)
bellofshn.head()
```

```python
print(len(bellofshn))
print(erate)
```

```python
rc = len(bellofshn)
price_dollars = []
for i in range(rc):
    if bellofshn.PaidCurrency[i] == 'Euro':
        price_dollars.append(bellofshn.Price_USD[i] * erate[0]*85)
        
    elif bellofshn.PaidCurrency[i] == 'Pound':
        price_dollars.append(bellofshn.Price_USD[i] * erate[1]*102)
        
    elif bellofshn.PaidCurrency[i] == 'Indian Rupee':
        price_dollars.append(bellofshn.Price_USD[i] * erate[2])
        
    else:
        price_dollars.append(bellofshn.Price_USD[i])
        
price_dollars
```

```python
bellofshn.Price_USD = price_dollars
bellofshn[['Price','PaidCurrency','Price_USD']]
```

```python
bellofshn['PriceUSDRound'] = bellofshn.Price_USD.round(2)
```
