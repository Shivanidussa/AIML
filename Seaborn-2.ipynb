```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\Shivani Dussa\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(9,4))
sns.boxplot(data = df,x = 'writing score',y='lunch')
```

```python
plt.figure(figsize=(11,5))
sns.boxplot(data = df,x= 'writing score',y='parental level of education',hue = 'test preparation course',palette = 'Set2')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(15,8))
sns.violinplot(data = df,x= 'writing score',y='parental level of education',hue = 'test preparation course',
              split = True,inner = 'stick',bw = 0.1)
plt.legend(bbox_to_anchor=(1.25,1))
```

```python
plt.figure(figsize=(11,5))
sns.swarmplot(data = df,x= 'reading score',y='gender',size = 2,hue = 'test preparation course',palette = 'Set1')
plt.legend(bbox_to_anchor=(1.25,1))
```

```python
plt.figure(figsize=(11,5))
sns.boxenplot(data = df,x= 'writing score',y='lunch',hue = 'test preparation course',palette = 'Set1')
plt.legend(bbox_to_anchor=(1.25,1))
```

Comparision plots 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\Shivani Dussa\\Downloads\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data = df,x = 'parental level of education', y = 'reading score', kind = 'scatter')
```

```python
sns.jointplot(data = df,x = 'writing score', y = 'reading score', kind = 'kde',alpha = 1,shade = True)
```

```python
sns.jointplot(data = df,x = 'writing score', y = 'reading score', hue = 'race/ethnicity')
plt.legend(bbox_to_anchor =(1.5,1))
```

```python
sns.pairplot(data = df)
```

```python
sns.pairplot(data = df,hue = 'race/ethnicity',diag_kind = 'hist')
```

```python
sns.pairplot(data = df,hue = 'race/ethnicity',diag_kind = 'hist',corner = True)
```

Grid Plots

```python
sns.catplot(data = df,x = 'race/ethnicity',y = 'math score')
```

```python
sns.catplot(data = df,x = 'race/ethnicity',y = 'math score',kind = 'box',col= 'lunch',row = 'test preparation course',
           hue = 'parental level of education')
```

```python
sns.catplot(data = df,x = 'race/ethnicity',y = 'math score',kind = 'violin',col= 'lunch',row = 'test preparation course')
```

```python
s = sns.PairGrid(df,hue = 'gender')

s = s.map_upper(sns.scatterplot)

s = s.map_diag(sns.histplot)

s = s.map_lower(sns.kdeplot)

s = s.add_legend()
```
